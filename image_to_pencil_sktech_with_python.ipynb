# **Data science task4:image to pencil sktech with python**

# **importing libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from skimage import io
```

# **Loading the Image**

```python
img = io.imread("https://i.pinimg.com/originals/c5/11/f0/c511f0da93cdb1c1db678ee8d8b587ef.jpg")
```

```python
cv2_imshow(img)
```

```python
gray_filter = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_filter)
```

```python
inverted_filter = cv2.bitwise_not(gray_filter)
cv2_imshow(inverted_filter)
```

```python
image_smoothing = cv2.GaussianBlur(inverted_filter, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_filter, image_smoothing)
cv2_imshow(pencil_sketch)
```
