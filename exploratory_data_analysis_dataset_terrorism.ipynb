# **exploratory data analysis dataset-terrorism**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r'/content/global terrorism.csv',encoding="latin")
```

```python
df1 = pd.DataFrame(df, df.rename(columns={ 'success': 'Success', 'location': 'Location', 'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'country': 'Country_code', 'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True))
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
d=df1['City'].value_counts().to_frame().drop('Unknown').sort_values('City',axis=0,ascending=False).head(10)
d.plot(kind='bar',figsize=(20,10),color='yellow')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Total attacks",fontsize=15)
plt.title("Top 10 most affected cities",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
index1=df1['Country'].value_counts()[:10].index
values1=df1['Country'].value_counts()[:10].values
sns.barplot(index1,values1,palette='rocket')
plt.title('Top 10 most affected countries', fontsize = 20)
plt.xlabel('Country', fontsize = 15)
plt.ylabel('Total attacks', fontsize = 15)
plt.xticks(rotation = 50)
plt.show()
```

```python
d1=df1['Region'].value_counts().to_frame().sort_values('Region',axis=0,ascending=False).head(10)
d1.plot(kind='bar',figsize=(20,10),color='black')
plt.xticks(rotation = 50)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Total attacks",fontsize=15)
plt.title("Top 10 most affected regions",fontsize=20)
plt.show()
```

```python
d2=df1[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10)
d2.plot(kind='bar',color='yellow',figsize=(20,10))
plt.title("Top 10 terrorist attack groups",fontsize=20)
plt.xlabel("Terrorist group",fontsize=20)
plt.ylabel("Total attacks",fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

```python
df1['AttackType'].value_counts().plot(kind='bar',figsize=(20,10),color='brown')
plt.xticks(rotation = 50)
plt.xlabel("Attack Type",fontsize=15)
plt.ylabel("Total attacks",fontsize=15)
plt.title("Most used attack types",fontsize=20)
plt.show()
```
